```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
import seaborn as sns
```

```python
df['Date Joined']=pd.to_datetime(df['Date Joined'])
df['Year Founded']=pd.to_datetime(df['Year Founded'])


```

```python
display(df['Valuation'].value_counts())
```

```python
display(df['Funding'].value_counts())
```

```python
df['Valuation']=df['Valuation'].str.replace('Unknown','0')
df['Valuation']=df['Valuation'].str.replace('B','000000000')
df['Valuation']=df['Valuation'].str.replace('M','000000')
```

```python
df['Valuation']=pd.to_numeric(df['Valuation'])
```

```python
df['Valuation']=df['Valuation'].str.replace(',','')
df['Valuation']=df['Valuation'].str.replace('$',' ')
```

```python
df['Valuation']=pd.to_numeric(df['Valuation'])
```

```python
df['Funding']=df['Funding'].str.replace(',','')
df['Funding']=df['Funding'].str.replace('$',' ')
df['Funding']=df['Funding'].str.replace('Unknown','0')
df['Funding']=df['Funding'].str.replace('B','000000000')
df['Funding']=df['Funding'].str.replace('M','000000')
```

```python
df['Funding']=pd.to_numeric(df['Funding'])
```

```python
df.info()
```

```python
df['month']=df['Date Joined'].dt.month
df.head()
```

```python
df['month_txt']=df['Date Joined'].dt.month_name()
df.head()
```

```python
df = df.drop('month', axis=1)
```

```python
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
df_by_month=df.groupby(['month_txt'])[['Valuation', 'Funding']].sum().reindex(month_order)
```

```python
plt.bar(x=df_by_month.index,height=df_by_month['Valuation'],label='valuation')
plt.title(' Total valuation per month')
plt.xlabel('Months')
plt.ylabel('valuation (in $) ')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
year=df['Date Joined'].dt.year
df['year']=year
df.head()
```

```python
df_by_year=df.groupby(['year'])[['Valuation', 'Funding']].sum()
df_by_year
```

```python
plt.bar(x=df_by_year.index,height=df_by_year['Valuation'],label='valuation')
plt.title(' Total valuation per year')
plt.xlabel('Years')
plt.ylabel('valuation (in $) ')
```

```python
plt.bar(x=df_by_year.index,height=df_by_year['Funding'],label='funding')
plt.title(' Total funding per year')
plt.xlabel('Years')
plt.ylabel('funding (in $) ')
```

```python
df.groupby(['Country/Region'])[['Valuation', 'Funding']].sum().sort_values(by='Valuation',ascending=False).head(10)
```

```python
 Industry_summary =df.groupby(['Industry'])[['Valuation', 'Funding']].sum().sort_values(by='Valuation',ascending=False).head(10)
```

```python
plt.bar(x=Industry_summary.index,height=Industry_summary['Valuation'], label='valuation')
plt.xlabel('Industry')
plt.ylabel('Valuation (in $) ')
plt.title('Total valuation per industry')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
plt.bar(x=Industry_summary.index ,height=Industry_summary['Funding'], label='funding')
plt.xlabel('Industry')
plt.ylabel('Funding (in $) ')
plt.title('Total funding per industry')
plt.xticks(rotation=90)
```
