
 **Categorizing vote_average coloumn**
Categorizing  movies as average average ,below_average,not_popular,popular


```python
def categorize_col(df,col,labels):
  edges=(df[col].describe()['min'],
         df[col].describe()['25%'],
         df[col].describe()['50%'],
         df[col].describe()['75%'],
         df[col].describe()['max'])
  df[col]=pd.cut(df[col],edges,labels=labels,duplicates='drop') #cuttng the coloumn as mentioned in edges and replacing values as in labels
  return df
```

```python
labels=['not_popular','below_average','average','popular']
df=categorize_col(df,'Vote_Average',labels)
```
