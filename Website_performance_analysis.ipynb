```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/data-export (1).csv")
```

```python
df.head(10)
```

```python
df.columns=df.iloc[0]
df.head()
```

```python
#df=df.drop(index=0)
df.head()
```

```python
df.columns=['channel group','datehour','users','Sessions',	'Engaged sessions',	'Average engagement time per session',	'Engaged sessions per user',	'Events per session',	'Engagement rate',	'Event count']
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df['datehour']=pd.to_datetime(df['datehour'],format='%Y%m%d%H', errors="coerce")
```

```python

numeric_cols=df.columns.drop(['channel group','datehour'])
df[numeric_cols]=df[numeric_cols].apply(pd.to_numeric)
```

```python
df["hour"]=df['datehour'].dt.hour
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.set(style='whitegrid')
```

```python
plt.figure(figsize=(10,6))
df.groupby("datehour")[['Sessions','user']].sum().plot(ax=plt.gca())
plt.xlabel("Datehour")
plt.ylabel("Session count")
plt.title("Sessions and users over time ")
plt.show()
```

# Total Users by channel

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df,x='channel group',y='users',estimator=np.sum ,hue='channel group',palette='viridis', legend=False)
plt.xlabel("Channel")
plt.ylabel("Users")
plt.title("Total users by channel")
plt.xticks(rotation=90)
plt.show()
```

```python
df.head()
```

#average engagement time per channel

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df,x='channel group',y='Average engagement time per session',estimator=np.mean ,hue='channel group',palette='magma', legend=False)
plt.xlabel("Channel")
plt.ylabel("Average engagement time")
plt.title("Average engagement time per channel")
plt.xticks(rotation=90)
```

#Engagement rate distribution by channel

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df,x='channel group',y='Engagement rate',palette='coolwarm')
plt.xlabel("Channel")
plt.ylabel("Engagement rate")
plt.title("Engagement rate distribution by channel")
plt.xticks(rotation=90)
```

#Engaged V/S non engaged sessions

```python
session_df=df.groupby("channel group")[['Sessions','Engaged sessions']].sum().reset_index()
session_df['non engaged sessions']=(session_df['Sessions']-session_df['Engaged sessions'])
sesions_df_melted=session_df.melt(id_vars='channel group',value_vars=['Engaged sessions','non engaged sessions'])
sns.barplot(data=sesions_df_melted,x='channel group',y='value',hue='variable',palette='viridis')
plt.xlabel("Channel")
plt.ylabel("Session count")
plt.title("Engaged V/S non engaged sessions")
plt.xticks(rotation=45)
```

# Traffic by hour code

```python
heat_map_data=df.groupby(['hour','channel group'])["Sessions"].sum().unstack().fillna(0)
plt.figure(figsize=(12,6))
sns.heatmap(heat_map_data,cmap="YlGnBu",linewidths=2,annot=True,fmt=' 0.0f', annot_kws={'size': 10})
plt.title("Traffic by hour code")
plt.xlabel("Channel")
plt.ylabel("Hour")
plt.show()
```

# engagement rate V/s sessions

```python
df_plot=df.groupby('datehour')[['Engagement rate','Sessions']].mean().reset_index()
plt.figure(figsize=(12,6))
sns.lineplot(data=df_plot,x='datehour',y='Engagement rate',label='Engagement rate')
sns.lineplot(data=df_plot,x='datehour',y='Sessions',label='Sessions')
plt.xlabel("Datehour")
plt.ylabel("Engagement rate/Sessions")
plt.title("Engagement rate V/s sessions")
plt.xticks(rotation=45)
```
